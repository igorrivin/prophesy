```python
from prostrat import *
```

```python
sptab = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
```

```python
symbols= list(sptab.Symbol)
```

```python
preddict = {}
```

```python
for i in symbols:
    try:
        preddict[i] = doprof(Stock(i, 'SMART', 'USD'))
    except:
        pass
```

```python
doprof(Stock('MMM', 'SMART', 'USD'))
```

```python
contract = Stock('MMM', 'SMART', 'USD')
```

```python
bars = ib.reqHistoricalData(
        contract,
        endDateTime='',
        durationStr='50 D',
        barSizeSetting='1 hour',
        whatToShow='TRADES',
        useRTH=True,
        formatDate=1)
```
