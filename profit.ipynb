```python
from ib_insync import *
import pandas as pd
import numpy as np
util.startLoop()

ib = IB()
ib.connect('127.0.0.1', 7497, clientId=15)
```

```python
from fbprophet import Prophet
```

```python
contract = Future('ES', '201906', 'GLOBEX')
#contract = Stock('GE', 'SMART', 'USD')

ib.reqHeadTimeStamp(contract, whatToShow='TRADES', useRTH=True)
```

```python
bars = ib.reqHistoricalData(
        contract,
        endDateTime='',
        durationStr='50 D',
        barSizeSetting='1 hour',
        whatToShow='TRADES',
        useRTH=True,
        formatDate=1)
```

```python
df = util.df(bars)
```

```python
df.head()
```

```python
prodf = df[['date', 'average']]
```

```python
prodf.columns = ['ds', 'y']
```

```python
prodf.y = np.log(prodf.y)
```

```python
#prodf.reset_index(inplace=True)
```

```python
pp = Prophet(changepoint_prior_scale=0.001)
```

```python
pp.fit(prodf)
```

```python
future = pp.make_future_dataframe(24, freq='1H')
```

```python
forecast = pp.predict(future)
```

```python
ff = forecast[['ds', 'yhat']]
```

```python
both = pd.concat([prodf, ff], axis=1)
```

```python
both['pdiff'] = both['yhat'] - both['y']
both['rdiff'] = -both.y.diff(-24)
both['rdiff2'] = -both.y.diff(-7)
```

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.regplot(x='pdiff', y='rdiff2', data=both)
```

```python
both.describe()
```
